```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def SampleFromUnitSphere(d):
    z = np.random.normal(size=d)
    return z/np.linalg.norm(z)

def SampleNFromUnitSphere(n, d):
    z = np.random.normal(size=(n,d))
    return (z.T/np.linalg.norm(z, axis=1)).T
```

```python
# Part A
n = 100
dList = [2,5,10,20,100]

for d in dList:
    dist = []
    x = SampleNFromUnitSphere(n, d)
    
    for i in range(len(x)):
        for j in range(i+1, len(x)):
            dist.append(np.linalg.norm(x[i]-x[j]))
            
    plt.hist(dist, bins=20, range=(0,2))
    plt.xlabel('Interpoint Distance')
    plt.ylabel('Count')
    plt.title('Histogram of interpoint distances for d = '+str(d))
    plt.show()
    plt.savefig('hist'+str(d)+'.png')
    plt.close()
```

```python
# Part C
d = 1000
n = 10000
x = SampleNFromUnitSphere(n, d)
u = []
s = []

for i in range(1, len(x)):
    localDist = []
    for j in range(0, i):
        localDist.append(np.linalg.norm(x[i]-x[j]))
        
    u.append(np.max(localDist))
    s.append(np.min(localDist))
    
plt.plot(u)
plt.plot(s)
plt.ylim([0, 2])
plt.legend(['U - largest distance', 'S - smallest distance'])
plt.xlabel('Value of i')
plt.ylabel('Interpoint distance')
plt.title('Largest/Smallest Interpoint distance')
```
